```python
import pandas as pd
import numpy as np
import ast
```

```python
credits = pd.read_csv('credits.csv')
```

```python
meta = pd.read_csv('movies_metadata.csv')
```

```python
meta
```

```python
meta['release_date'] = pd.to_datetime(meta['release_date'], errors = 'coerce')
```

```python
meta
```

```python
meta['release_year'] = meta['release_date'].dt.year
```

```python
meta_new = meta.loc[meta['release_year']<=2017,['id','genres','title', 'release_year']]
```

```python
meta_new['id'] = meta_new['id'].astype('int')
```

```python
data = pd.merge(meta_new, credits, on = 'id')
```

```python
pd.set_option('display.max_colwidth', 75)
data
```

```python
data
```

```python
data['genres'] = data['genres'].map(lambda x: ast.literal_eval(x))
data['cast'] = data['cast'].map(lambda x: ast.literal_eval(x))
data['crew'] = data['crew'].map(lambda x: ast.literal_eval(x))
```

```python
data
```

```python
def make_list(x):
    gen = []
    st = " "
    for i in x:
        if i.get('name') == 'Scientific Fiction':
            scifi = 'Sci-Fi'
            gen.append(scifi)
        else:
            gen.append(i.get('name'))
    if gen == []:
        return np.nan
    else:
        return st.join(gen)
```

```python
data['genres_list'] = data['genres'].map(lambda x: make_list(x))
```

```python
data['genres_list']
```

```python
def get_actors1(x):
    cast = []
    for i in x:
        cast.append(i.get('name'))
    if cast == []:
        return np.nan
    else:
        return cast[0]
```

```python
data['actor_1_name'] = data['cast'].map(lambda x: get_actors1(x))
```

```python
data
```

```python
def get_actors2(x):
    cast = []
    for i in x:
        cast.append(i.get('name'))
    if cast == [] or len(cast)<=1:
        return np.nan
    else:
        return cast[1]
```

```python
data['actor_2_name'] = data['cast'].map(lambda x: get_actors2(x))
```

```python
def get_actors3(x):
    cast = []
    for i in x:
        cast.append(i.get('name'))
    if cast == [] or len(cast)<=2:
        return np.nan
    else:
        return cast[2]
```

```python
data['actor_3_name'] = data['cast'].map(lambda x: get_actors3(x))
```

```python
data
```

```python
data['crew'][0]
```

```python
def get_director(x):
    crew = []
    st = " "
    for i in x:
        if i.get('job') == 'Director':
            crew.append(i.get('name'))
    if crew == []:
        return np.nan
    else:
        return (st.join(crew))
```

```python
data['director'] = data['crew'].map(lambda x: get_director(x))
```

```python
data
```

```python
movies = data.loc[:,['title', 'genres_list', 'actor_1_name', 'actor_2_name', 'actor_3_name', 'director']]
movies
```

```python
movies.isna().sum()
```

```python
movies = movies.dropna(how = 'any')
```

```python
movies.isna().sum()
```

```python
movies['title'] = movies['title'].str.lower()
```

```python
movies['combined'] = movies['actor_1_name']+ ' '+movies['actor_2_name']+' '+movies['actor_3_name']+ ' '+movies['director']+ ' '+movies['genres_list']
```

```python
movies
```

```python
movies.drop_duplicates(subset = 'title',keep = 'last', inplace = True)
```

```python
movies
```

```python
movies.to_csv('final.csv', index = False)
```
