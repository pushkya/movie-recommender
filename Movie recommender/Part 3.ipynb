```python
from tmdbv3api import TMDb
import json
import requests
tmdb = TMDb()
tmdb.api_key = '863e2a8d5766d83647121e8e4650a4ff'
```

```python
tmdb_movie = Movie()
```

```python
def get_genres(x):
    genres = []
    result = tmdb_movie.search(x)
    movie_id = result[0].id
    response = requests.get('https://api.themoviedb.org/3/movie/{}?api_key={}'.format(movie_id,tmdb.api_key))
    data_json = response.json()
    if data_json['genres']:
        data_genres = ''
        for i in range(0, len(data_json['genres'])):
            genres.append(data_json['genres'][i]['name'])
        return data_genres.join(genres)
    else:
        return np.nan
```

```python
df['genres'] = df['Title'].map(lambda x: get_genres(str(x)))
```

```python
def get_director(x):
    if " (director)" in x:
        return x.split(" (director)")[0]
    elif " (directors)" in x:
        return x.split(" (directors)")[0]
    else:
        return x.split(" (director/screenplay)")[0]
```

```python
df_2018['director'] = df['Cast and crew'].map(lambda x: get_director(x))
```

```python
def get_actor1(x):
    return ((x.split("screenplay); ")[-1]).split(", ")[0])
```

```python
df_2018['actor1'] = df['Cast and crew'].map(lambda x: get_actor1(x))
```

```python
def get_actor2(x):
    if len((x.split(" screenplay); ")[-1]).split(", ")) < 2:
        return np.nan
    else:
        return ((x.split(" screenplay); ")[-1]).split(", ")[1])
```

```python
df_2018['actor2'] = df_2018['Cast and crew'].map(lambda x: get_actor2(x))
```

```python
def get_actor3(x):
    if len((x.split(" screenplay); ")[-1]).split(", ")) < 3:
        return np.nan
    else:
        return ((x.split(" screenplay); ")[-1]).split(", ")[2])
```
